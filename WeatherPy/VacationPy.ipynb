## Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# A max temperature lower than 27 degrees but higher than 21
criteria_ideal_weather = city_data_df[(city_data_df["Max Temp"] >21.00) & (city_data_df["Max Temp"] <27.00)] 
criteria_ideal_weather
```

```python
# Incorporate two more criteria: 
# Wind speed less than 4.5 m/s
# Zero cloudiness

criteria_ideal_weather = city_data_df[((city_data_df["Max Temp"] >21.00) & (city_data_df["Max Temp"] <27.00)) \
& (city_data_df["Wind Speed"] <4.5)\
& (city_data_df["Cloudiness"] == 0)]
criteria_ideal_weather

## PROBLEM!! my random cities.csv dataset produced no output because only 19 cities are within this temp range, and none have zero cloudiness. 
## So we will adjust to new specifications.  
```

```python
# New adjusted specifications: 
# A max temperature lower than 80 degrees but higher than 40
# Wind speed less than 4.5 m/s
# Zero cloudiness

combined_ideal_weather = city_data_df[((city_data_df["Max Temp"] >40.00) & (city_data_df["Max Temp"] <80.00))
& (city_data_df["Wind Speed"] <4.5) \
& (city_data_df["Cloudiness"] == 0)]
combined_ideal_weather
```

```python
# Narrow down cities that fit criteria and drop any results/rows with null values
combined_ideal_weather = combined_ideal_weather.dropna(how = 'any')  
# Display sample data
combined_ideal_weather
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = pd.DataFrame({
            "City": combined_ideal_weather ["City"],
            "Country": combined_ideal_weather ["Country"],
            "Lat": combined_ideal_weather ["Lat"],
            "Lng": combined_ideal_weather ["Lng"],
            "Humidity":combined_ideal_weather ["Humidity"]
           })
            
# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df["Hotel Name"] = ""

# Display sample data
hotel_df
```

```python
# Set parameters to search for a hotel
radius = 10000 
params = {"categories":"accommodation.hotel",
          "apiKey":geoapify_key,
          "limit":20
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # get latitude, longitude from the DataFrame
    longitude = row["Lng"]
    latitude = row["Lat"]
    
    # Add filter and bias parameters with the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{longitude},{latitude},{radius}"
    params["bias"] = f"proximity:{longitude},{latitude}"
    
    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"


    # Make and API request using the params dictionaty
    name_address = requests.get(base_url,params=params)
    
    # Convert the API response to JSON format
    name_address = name_address.json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"
      
    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
%%capture --no-display

# Configure the map plot
hotel_map = hotel_df.hvplot.points(
         "Lng", 
         "Lat", 
         geo=True, 
         size="Humidity", 
         color="City", 
         tiles="OSM", 
         frame_width = 700, 
         frame_height = 500,
         hover_cols = ["Hotel Name", "Country"]
)
# Display the map
hotel_map
```
